```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

n = 96
t = np.arange(n)
demand = np.sin(3 * np.pi * t / n) * np.linspace(1, 3, n)
pv = demand * 1.4
pv[:35] = 0
pv[60:] = 0
pv = np.convolve(pv, np.ones(10)/10, 'same')
shift = np.sin(3 * np.pi * t / n) * np.linspace(1, 3, n)
shift *= np.hstack([np.zeros(10), np.convolve(np.ones(25), np.ones(10)/10, 'same')*0.2,
                    np.convolve(np.ones(25), np.ones(10)/10, 'same')*0.2, np.zeros(36)])
changed_demand = demand -shift
ripple_control = np.zeros(n)
ripple_control[10:30] = 0.4

matplotlib.pyplot.xkcd()

plt.set_cmap('Set1')
colors = plt.get_cmap('Set1', 3)

fig, ax = plt.subplots(1, 1)

ax.plot(demand, label='demand')
ax.plot(changed_demand)
ax.plot(pv, label='PV generation')
ax.fill_between(t, demand, changed_demand, label='shifted demand')
ax.plot(ripple_control, label='control')
plt.legend(fontsize='small')
plt.xlabel('time [15 min]')
plt.ylabel('P [MW]')

```
